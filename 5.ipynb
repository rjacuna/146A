$\begin{cases}\frac{dx}{dt}= x(1.5 -x -0.5y)\\ \frac{dy}{dt} = y(2 -y -0.75x)\end{cases}$

```python
x,y = symbols('x,y');
F = x*(3/2 -1*x -1*y/2);
G = y*(2 -1*y -3*x/4); F,G
```

```python
solve(F)
```

```python
solve(G.subs(x,0))
```

```python
solve(G.subs(y,3-2*x))
```
